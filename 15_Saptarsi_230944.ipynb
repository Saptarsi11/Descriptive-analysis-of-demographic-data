# Task 1: Describe the frequency distributions of the variables. Consider also the differences between the sexes and regions.

```python
fig, axes = plt.subplots(2, 3, figsize=(10, 10))


ax=sns.histplot(df2022['Life.Expectancy.at.Birth..Both.Sexes'],stat='density', ax=axes[0, 0])
bx=sns.histplot(df2022['Life.Expectancy.at.Birth..Females'],stat='density', ax=axes[0, 1])
cx=sns.histplot(df2022['Life.Expectancy.at.Birth..Males'],stat='density', ax=axes[0, 2])
dx=sns.histplot(df2022['Under.Age.5.Mortality..Both.Sexes'],stat='density', ax=axes[1, 0])
ex=sns.histplot(df2022['Under.Age.5.Mortality..Females'],stat='density', ax=axes[1, 1])
fx=sns.histplot(df2022['Under.Age.5.Mortality..Males'],stat='density', ax=axes[1, 2])


ax.set(xlabel = 'Life.Expectancy.at.Birth..Both.Sexes', ylim=(0.00, 0.08),xlim=(50,100))
bx.set(xlabel = 'Life.Expectancy.at.Birth..Females' , ylim=(0.00, 0.08),xlim=(50,100))
cx.set(xlabel = 'Life.Expectancy.at.Birth..Males', ylim=(0.00, 0.08),xlim=(50,100))
dx.set(xlabel = 'Under.Age.5.Mortality..Both.Sexes', ylim=(0.00, 0.05),xlim=(0,200))
ex.set(xlabel = 'Under.Age.5.Mortality..Females' , ylim=(0.00, 0.05),xlim=(0,200))
fx.set(xlabel = 'Under.Age.5.Mortality..Males', ylim=(0.00, 0.05),xlim=(0,200))
#plt.savefig('Figure 1.pdf')
```

```python
a=plt.scatter(df2022['Life.Expectancy.at.Birth..Both.Sexes'], df2022['Life.Expectancy.at.Birth..Females'], color='r')
b=plt.scatter(df2022['Life.Expectancy.at.Birth..Both.Sexes'], df2022['Life.Expectancy.at.Birth..Males'], color='b')
plt.xlabel('Life Expectancy Range',size=12)
plt.ylabel('Life Expectancy Range',size=12)
plt.title('Differences between the sexes')
plt.legend((a,b),('Females','Males'))
plt.show()
```
